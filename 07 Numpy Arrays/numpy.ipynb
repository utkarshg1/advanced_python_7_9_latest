# Arrays instead of List for mathematical calculations

### Multipying by 3 to a list vs array

```python
e = [12, 13, 14, 15, 16]
type(e)
```

```python
# in list * operator means repeat list
e * 3
```

```python
e1 = []
for i in e:
    print(f"Multiplying {i} by 3 = {i*3}")
    e1.append(i * 3)
```

```python
e1
```

```python
e2 = np.array(e)
e2
```

```python
type(e2)
```

```python
e3 = e2 * 3
print(e3)
```

```python
e2
```

```python
e2 + 5
```

```python
3 * e2 + 5
```

# broadcasting in numpy array -> Any mathematical operation on arrays will be applied to individual elements in array

```python
e2
```

```python
e2**2
```

```python
e2**3
```

```python
e2 ** (1 / 2)
```

```python
np.sqrt(e2)
```

# Taking product of elements in list vs array

```python
g = [2, 3, 4, 5, 6]
type(g)
```

```python
2 * 3 * 4 * 5 * 6
```

```python
# Intialize product by 1
p = 1
# Apply for loop on g
for i in g:
    p = p * i
```

```python
print(p)
```

```python
g1 = np.array(g)
g1
```

```python
type(g1)
```

```python
g1.prod()
```

```python
g1.sum()
```

```python
g1.mean()
```

```python
g1.std()
```

```python
g1.max()
```

```python
g1.min()
```

### Indexing


```python
# Index of highest value
g1.argmax() 
```

```python
# Index of lowest value
g1.argmin()
```
