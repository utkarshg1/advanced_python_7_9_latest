# Write a webscraping class to scrape any wikipedia page

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd


class WikiScraper:

    def __init__(self, url: str):
        self.url = url
        self.home_page = "https://en.wikipedia.org"
        response = requests.get(self.url)
        response.raise_for_status()
        self.soup = BeautifulSoup(response.content)

    def get_title(self):
        title_tag = self.soup.find("title")
        return title_tag.text

    def get_h1_tag(self):
        h1_tag = self.soup.find("h1", class_="firstHeading")
        return h1_tag.text

    def get_subheadings(self):
        sub_tags = self.soup.find_all("div", class_="mw-heading")
        sub_text = [tag.text for tag in sub_tags]
        return sub_text

    def get_paras(self):
        p_tags = self.soup.find_all("p")
        p_text = [tag.text for tag in p_tags]
        return p_text

    def get_image_urls(self):
        a_tags = self.soup.find_all("a", class_="mw-file-description")
        image_urls = [self.home_page + tag.get("href") for tag in a_tags]
        return image_urls

    def get_tables(self):
        table_tags = self.soup.find_all("table", class_="wikitable")
        dfs = []
        for tag in table_tags:
            d = pd.read_html(str(tag))[0]
            dfs.append(d)
        return dfs
```

# I have 5 wikipedia urls get all the images from these urls

```python
urls = [
    "https://en.wikipedia.org/wiki/World_population",
    "https://en.wikipedia.org/wiki/Data_science",
    "https://en.wikipedia.org/wiki/Data_analysis",
    "https://en.wikipedia.org/wiki/Python_(programming_language)",
    "https://en.wikipedia.org/wiki/Rust_(programming_language)",
]
print(urls)
```

```python
images = {}

for i in urls:
    scraper = WikiScraper(url=i)
    print(f"Getting image urls from {scraper.url}")
    h1_text = scraper.get_h1_tag()
    print(h1_text)
    img_urls = scraper.get_image_urls()
    print(img_urls)
    images[h1_text] = img_urls
```

```python
import json

with open("image.json", "w") as f:
    json.dump(images, f, indent=4)
```
