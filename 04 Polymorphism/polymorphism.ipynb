# Assignment
Assignment Create a class Triangle to do following
1. attributes 3 side lengths a, b, c
2. a function to check that triangle forms or not
3. a function to calculate perimeter
4. to calulate area with herons formula

```python
from dataclasses import dataclass
import math
```

```python
@dataclass
class Triangle:
    a: int | float
    b: int | float
    c: int | float

    def __post_init__(self):
        if self.a <= 0 or self.b <= 0 or self.c <= 0:
            raise ValueError("Sides of triangle cannot be negative")
        if not self.is_triangle():
            raise ValueError(
                "Given sides do not form a triangle, Sum of two sides should be greater than third"
            )

    def is_triangle(self) -> bool:
        return (
            (self.a + self.b > self.c)
            and (self.b + self.c > self.a)
            and (self.c + self.a > self.b)
        )

    def perimeter(self) -> int | float:
        return self.a + self.b + self.c

    def area(self) -> float:
        s = self.perimeter() / 2
        a = math.sqrt(s * (s - self.a) * (s - self.b) * (s - self.c))
        return a
```

# Apply polymorphism for calculating shape area and perimeter
1. Rectangle
2. Circle

```python
@dataclass
class Rectangle:
    length: int | float
    height: int | float

    def perimeter(self) -> int | float:
        p = 2 * (self.length + self.height)
        return p

    def area(self) -> int | float:
        a = self.length * self.height
        return a
```

```python
import math

math.pi
```

```python
@dataclass
class Circle:
    radius: int | float

    def perimeter(self):
        p = 2 * math.pi * self.radius
        return p

    def area(self):
        a = math.pi * (self.radius**2)
        return a
```

```python
r1 = Rectangle(length=12, height=20)
print(r1)
print(type(r1))
print(r1.perimeter())
print(r1.area())
```
